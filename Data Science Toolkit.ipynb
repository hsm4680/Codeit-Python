- 불린 인덱싱 : 특정 조건을 충족하는 값을 가져옴

```python
gdp_array = np.array([6610, 7637, 8127, 8885, 10385, 12565, 13403, 12398, 8282, 10672])
```

```python
gdp_array > 10000 #10000보다 큰 인덱스에는 true, 아닌 인덱스에는 false가 들어감
```

```python
mask = gdp_array>10000
gdp_array[mask] #true가 있는 인덱스들만 가져옴
```

```python
gdp_array[gdp_array>10000]
```

```python
(gdp_array > 10000) | (gdp_array < 8000) #OR 연산
```

```python
gdp_array[(gdp_array > 10000) | (gdp_array < 8000)]
```

```python
(gdp_array <= 10000) & (gdp_array >= 8000) #AND 연산
```

# Matplotlib과 데이터 시각화

```python
import matplotlib.pyplot as plt
```

```python
year_array = np.array([2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])

stock_array = np.array([
    14.46, 19.01, 20.04, 27.59, 26.32,
    28.96, 42.31, 39.44, 73.41, 132.69
])
```

```python
plt.plot(year_array, stock_array) #x축과 y축에 들어갈 값을 순차적으로 넣으면 됨
plt.show()
```

```python
name_array = np.array(['mark', 'dongwook', 'hyojune', 'sowon', 'taeho'])
votes_array = np.array([5, 10, 6, 8, 3])
```

```python
plt.bar(name_array, votes_array) #막대그래프
plt.show()
```

```python
height_array = np.array([
    165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
    165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
    172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

weight_array = np.array([
    62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
    57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
    67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])

```

```python
plt.scatter(height_array, weight_array) #산점도 그래프 : 두 변수의 연관성 파악
```

```python
plt.scatter(height_array, weight_array, c = 'Green') #c 파라미터는 컬러를 의미, 컬러코드도 가능함
plt.title('Height and Weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
```

```python
plt.scatter(height_array, weight_array, c = 'Green', marker='+') #c 파라미터는 컬러를 의미, 컬러코드도 가능함
plt.title('Height and Weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
```

```python
plt.scatter(height_array, weight_array, c = 'Green', marker='s') #c 파라미터는 컬러를 의미, 컬러코드도 가능함
plt.title('Height and Weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
```

```python
plt.figure(figsize=(10, 4))  #해당 설정에서만 플랏 그래프의 크기가 변경됨
plt.scatter(height_array, weight_array)
plt.title('Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (5, 5) #해당 노트북 안에서 플랏의 크기가 고정됨
plt.scatter(height_array, weight_array)
plt.title('Scatter Plot')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
plt.scatter(height_array, weight_array)
plt.rc('font', family='AppleGothic') #한글의 경우 폰트를 지정해주어야 그래프에서 깨지지 않음
plt.title('키와 몸무게')
plt.xlabel('키(cm)')
plt.ylabel('몸무게(kg)')
plt.show()
```
